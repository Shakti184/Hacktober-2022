```python
# Recommendation System

moviemat = df.pivot_table(index='user_id', columns='title', values='rating')
```

```python
moviemat
```

```python
ratings.sort_values('num of rating', ascending =False).head()
```

### Predict Movie

```python
def predict_movies(movie_name):
    movie_user_ratings = moviemat[movie_name]
    similar_to_movie = moviemat.corrwith(movie_user_ratings)
    corr_of_movie = pd.DataFrame(similar_to_movie, columns = ['Correlation'])
    corr_of_movie.dropna(inplace=True)
    corr_of_movie = corr_of_movie.join(ratings['num of rating'])
    predictions = corr_of_movie[corr_of_movie['num of rating']>100].sort_values('Correlation', ascending = False)
    return predictions
```

```python
prediction = predict_movies('Titanic (1997)')
```
